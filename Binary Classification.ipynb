Context
This data set dates from 1988 and consists of four databases: Cleveland, Hungary, Switzerland, and Long Beach V. It contains 76 attributes, including the predicted attribute, but all published experiments refer to using a subset of 14 of them. The "target" field refers to the presence of heart disease in the patient. It is integer valued 0 = no disease and 1 = disease.

Content
Attribute Information:

age -age	

sex - sex

cp - chest pain type (4 values)

chol - trestbps - resting blood pressure

fbs - serum cholestoral in mg/dl

restecg - fasting blood sugar > 120 mg/dl

thalach - resting electrocardiographic results (values 0,1,2)

exang - maximum heart rate achieved

oldpeak - exercise induced angina

oldpeak = ST depression induced by exercise relative to rest

slope - the slope of the peak exercise ST segment

ca - number of major vessels (0-3) colored by flourosopy

thal: 0 = normal; 1 = fixed defect; 2 = reversable defect

target - target

The names and social security numbers of the patients were recently removed from the database, replaced with dummy values.

```python
from sklearn.model_selection import train_test_split
X=df.drop('target',axis=1)
y=df.target

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
y_test.shape
```

```python
X_test.shape
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print('the accuracy of the model is : ', round(accuracy_score(y_test,y_pred),2))
```

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
model=Sequential()
model.add(Dense(11,activation='relu',input_dim=X_train.shape[1]))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

model.summary()
```

```python
model.fit(X_train,y_train,epochs=300)
```

```python
y_pred_nn=model.predict(X_test)
```

```python
model=Sequential()
model.add(Dense(11,activation='relu',input_dim=X_train.shape[1]))
model.add(Dense(5,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

model.summary()

model.fit(X_train,y_train,epochs=300)
```

```python
y_pred_nn1=model.predict(X_test)
y_pred_n2=[round(x[0]) for x in y_pred_nn1]
print('The accuracy of the deep neural net is :', round(accuracy_score(y_test,y_pred_n2),2))
```

```python
model=Sequential()
model.add(Dense(11,activation='relu',input_dim=X_train.shape[1]))
model.add(Dense(5,activation='relu'))
model.add(Dense(5,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

model.summary()

model.fit(X_train,y_train,epochs=300)
```

```python
y_pred_nn2=model.predict(X_test)
y_pred_n3=[round(x[0]) for x in y_pred_nn2]
print('The accuracy of the deep neural net is :', round(accuracy_score(y_test,y_pred_n3),2))
```
